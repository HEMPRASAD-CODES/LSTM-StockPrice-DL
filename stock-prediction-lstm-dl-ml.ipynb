```python
# 🧠 End-to-End Stock Price Prediction with Deep Learning
# 📈 LSTM Model for ADANIPORTS Stock Prediction
# 🎯 Objective: Predict future stock prices from raw historical data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

print("🚀 Starting End-to-End Stock Price Prediction")
print("=" * 50)

# 📊 STEP 1: DATA LOADING AND EXPLORATION
print("📊 STEP 1: Loading and Exploring Data")
print("-" * 30)

# Load the dataset
df = pd.read_csv('/kaggle/input/adaniports/ADANIPORTS.csv')

print(f"Dataset Shape: {df.shape}")
print(f"Date Range: {df['Date'].min()} to {df['Date'].max()}")
print("\nDataset Info:")
print(df.info())
print("\nFirst 5 rows:")
print(df.head())

# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date').reset_index(drop=True)

# Check for missing values
print(f"\nMissing Values:\n{df.isnull().sum()}")

# 📈 STEP 2: DATA VISUALIZATION
print("\n📈 STEP 2: Data Visualization")
print("-" * 30)

# Create comprehensive visualizations
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('ADANIPORTS Stock Analysis', fontsize=16, fontweight='bold')

# Price trend over time
axes[0, 0].plot(df['Date'], df['Close'], color='blue', linewidth=1)
axes[0, 0].set_title('Close Price Over Time')
axes[0, 0].set_xlabel('Date')
axes[0, 0].set_ylabel('Close Price (₹)')
axes[0, 0].grid(True)

# Volume trend
axes[0, 1].plot(df['Date'], df['Volume'], color='orange', linewidth=1)
axes[0, 1].set_title('Trading Volume Over Time')
axes[0, 1].set_xlabel('Date')
axes[0, 1].set_ylabel('Volume')
axes[0, 1].grid(True)

# Price distribution
axes[1, 0].hist(df['Close'], bins=50, color='green', alpha=0.7)
axes[1, 0].set_title('Close Price Distribution')
axes[1, 0].set_xlabel('Close Price (₹)')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].grid(True)

# High-Low spread
axes[1, 1].plot(df['Date'], df['High'] - df['Low'], color='red', linewidth=1)
axes[1, 1].set_title('Daily High-Low Spread')
axes[1, 1].set_xlabel('Date')
axes[1, 1].set_ylabel('Price Spread (₹)')
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()

# Show key statistics
print(f"\nKey Statistics for Close Price:")
print(f"Mean: ₹{df['Close'].mean():.2f}")
print(f"Std: ₹{df['Close'].std():.2f}")
print(f"Min: ₹{df['Close'].min():.2f}")
print(f"Max: ₹{df['Close'].max():.2f}")

# ⚙️ STEP 3: DATA PREPROCESSING
print("\n⚙️ STEP 3: Data Preprocessing")
print("-" * 30)

# Select features for the model (raw data approach)
# We'll use: Open, High, Low, Close, Volume
feature_columns = ['Open', 'High', 'Low', 'Close', 'Volume']
data = df[feature_columns].values

print(f"Selected Features: {feature_columns}")
print(f"Data Shape: {data.shape}")

# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Create sequences for LSTM (sliding window approach)
def create_sequences(data, sequence_length, target_column_index):
    """
    Create sequences for LSTM training
    Args:
        data: scaled data array
        sequence_length: number of time steps to look back
        target_column_index: index of the target column (Close price)
    Returns:
        X: input sequences, y: target values
    """
    X, y = [], []
    for i in range(sequence_length, len(data)):
        X.append(data[i-sequence_length:i])  # All features for sequence
        y.append(data[i, target_column_index])  # Only close price as target
    return np.array(X), np.array(y)

# Hyperparameters
SEQUENCE_LENGTH = 30  # Use 60 days of data to predict next day
TARGET_COLUMN = 3  # Close price is at index 3

print(f"Sequence Length: {SEQUENCE_LENGTH} days")

# Create sequences
X, y = create_sequences(scaled_data, SEQUENCE_LENGTH, TARGET_COLUMN)

print(f"Sequences Shape: X={X.shape}, y={y.shape}")

# Train-Test Split (80-20)
split_index = int(0.8 * len(X))
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

print(f"Training Set: X={X_train.shape}, y={y_train.shape}")
print(f"Test Set: X={X_test.shape}, y={y_test.shape}")

# 🤖 STEP 4: MODEL ARCHITECTURE
print("\n🤖 STEP 4: Building LSTM Model")
print("-" * 30)

def build_lstm_model(input_shape):
    """
    Build LSTM model architecture
    """
    model = Sequential([
        # First LSTM layer with dropout
        LSTM(units=50, return_sequences=True, input_shape=input_shape),
        Dropout(0.2),
        
        # Second LSTM layer with dropout
        LSTM(units=50, return_sequences=True),
        Dropout(0.2),
        
        # Third LSTM layer with dropout
        LSTM(units=50, return_sequences=False),
        Dropout(0.2),
        
        # Dense layers for final prediction
        Dense(units=25, activation='relu'),
        Dense(units=1, activation='linear')  # Linear for regression
    ])
    
    return model

# Build the model
model = build_lstm_model((X_train.shape[1], X_train.shape[2]))

# Compile the model
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='mse',  # Mean Squared Error for regression
    metrics=['mae']  # Mean Absolute Error
)

print("Model Architecture:")
model.summary()

# 🎯 STEP 5: MODEL TRAINING
print("\n🎯 STEP 5: Training the Model")
print("-" * 30)

# Callbacks for better training
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=0.0001,
        verbose=1
    )
]

# Train the model
print("Starting training...")
history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=32,
    validation_split=0.2,
    callbacks=callbacks,
    verbose=1
)

print("✅ Training completed!")

# 📊 STEP 6: TRAINING VISUALIZATION
print("\n📊 STEP 6: Training Performance Visualization")
print("-" * 30)

# Plot training history
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Loss curves
ax1.plot(history.history['loss'], label='Training Loss', color='blue')
ax1.plot(history.history['val_loss'], label='Validation Loss', color='red')
ax1.set_title('Model Loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss (MSE)')
ax1.legend()
ax1.grid(True)

# MAE curves
ax2.plot(history.history['mae'], label='Training MAE', color='blue')
ax2.plot(history.history['val_mae'], label='Validation MAE', color='red')
ax2.set_title('Model MAE')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Mean Absolute Error')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()

# 🔮 STEP 7: PREDICTIONS
print("\n🔮 STEP 7: Making Predictions")
print("-" * 30)

# Make predictions on test set
y_pred_scaled = model.predict(X_test, verbose=0)

# Create arrays for inverse scaling
# We need to create full feature arrays to inverse transform
def inverse_transform_predictions(scaler, predictions, target_column_index, num_features):
    """
    Inverse transform predictions to original scale
    """
    # Create dummy array with same shape as original features
    dummy = np.zeros((len(predictions), num_features))
    dummy[:, target_column_index] = predictions.flatten()
    
    # Inverse transform
    dummy_inverse = scaler.inverse_transform(dummy)
    
    return dummy_inverse[:, target_column_index]

# Inverse transform predictions and actual values
y_pred = inverse_transform_predictions(scaler, y_pred_scaled, TARGET_COLUMN, len(feature_columns))
y_actual = inverse_transform_predictions(scaler, y_test.reshape(-1, 1), TARGET_COLUMN, len(feature_columns))

print(f"Predictions Shape: {y_pred.shape}")
print(f"Actual Values Shape: {y_actual.shape}")

# 📈 STEP 8: EVALUATION METRICS
print("\n📈 STEP 8: Model Evaluation")
print("-" * 30)

# Calculate metrics
mse = mean_squared_error(y_actual, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_actual, y_pred)

# Calculate percentage error
mape = np.mean(np.abs((y_actual - y_pred) / y_actual)) * 100

print(f"🎯 Model Performance Metrics:")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): ₹{rmse:.2f}")
print(f"Mean Absolute Error (MAE): ₹{mae:.2f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")

# 📊 STEP 9: RESULTS VISUALIZATION
print("\n📊 STEP 9: Results Visualization")
print("-" * 30)

# Create comprehensive results visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('LSTM Model Prediction Results', fontsize=16, fontweight='bold')

# 1. Actual vs Predicted (Line Plot)
axes[0, 0].plot(y_actual, label='Actual Price', color='blue', linewidth=2)
axes[0, 0].plot(y_pred, label='Predicted Price', color='red', linewidth=2, alpha=0.8)
axes[0, 0].set_title('Actual vs Predicted Stock Prices')
axes[0, 0].set_xlabel('Test Sample')
axes[0, 0].set_ylabel('Stock Price (₹)')
axes[0, 0].legend()
axes[0, 0].grid(True)

# 2. Scatter Plot
axes[0, 1].scatter(y_actual, y_pred, alpha=0.6, color='green')
axes[0, 1].plot([y_actual.min(), y_actual.max()], [y_actual.min(), y_actual.max()], 'r--', lw=2)
axes[0, 1].set_title('Predicted vs Actual (Scatter)')
axes[0, 1].set_xlabel('Actual Price (₹)')
axes[0, 1].set_ylabel('Predicted Price (₹)')
axes[0, 1].grid(True)

# 3. Prediction Error Distribution
errors = y_actual - y_pred
axes[1, 0].hist(errors, bins=30, color='orange', alpha=0.7)
axes[1, 0].set_title('Prediction Error Distribution')
axes[1, 0].set_xlabel('Prediction Error (₹)')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].grid(True)

# 4. Recent Predictions (Last 50 samples)
recent_samples = min(50, len(y_actual))
axes[1, 1].plot(range(recent_samples), y_actual[-recent_samples:], 
                label='Actual', color='blue', linewidth=2)
axes[1, 1].plot(range(recent_samples), y_pred[-recent_samples:], 
                label='Predicted', color='red', linewidth=2, alpha=0.8)
axes[1, 1].set_title(f'Recent {recent_samples} Predictions')
axes[1, 1].set_xlabel('Sample')
axes[1, 1].set_ylabel('Stock Price (₹)')
axes[1, 1].legend()
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()

# 🔮 STEP 10: FUTURE PREDICTION
print("\n🔮 STEP 10: Predicting Next Day Price")
print("-" * 30)

# Use the last sequence to predict next day
last_sequence = X_test[-1:]  # Last sequence from test set
next_day_scaled = model.predict(last_sequence, verbose=0)

# Inverse transform
next_day_price = inverse_transform_predictions(scaler, next_day_scaled, TARGET_COLUMN, len(feature_columns))

print(f"📈 Next Day Predicted Price: ₹{next_day_price[0]:.2f}")

# Get the last actual price for comparison
last_actual_price = df['Close'].iloc[-1]
print(f"📊 Last Known Price: ₹{last_actual_price:.2f}")
print(f"📊 Predicted Change: ₹{next_day_price[0] - last_actual_price:.2f}")
print(f"📊 Predicted Change %: {((next_day_price[0] - last_actual_price) / last_actual_price) * 100:.2f}%")

# 📋 STEP 11: MODEL SUMMARY
print("\n📋 STEP 11: Final Model Summary")
print("=" * 50)

print("🎯 MODEL CONFIGURATION:")
print(f"  • Architecture: LSTM with 3 layers (50 units each)")
print(f"  • Sequence Length: {SEQUENCE_LENGTH} days")
print(f"  • Features Used: {len(feature_columns)} ({', '.join(feature_columns)})")
print(f"  • Training Samples: {len(X_train):,}")
print(f"  • Test Samples: {len(X_test):,}")

print("\n📊 PERFORMANCE SUMMARY:")
print(f"  • RMSE: ₹{rmse:.2f}")
print(f"  • MAE: ₹{mae:.2f}")
print(f"  • MAPE: {mape:.2f}%")

print("\n🔮 PREDICTION CAPABILITY:")
print(f"  • Next Day Price: ₹{next_day_price[0]:.2f}")
print(f"  • Model Confidence: Based on {SEQUENCE_LENGTH} days of historical data")

print("\n⚠️  IMPORTANT DISCLAIMERS:")
print("  • This model is for educational purposes only")
print("  • Stock market predictions are inherently uncertain")
print("  • Past performance does not guarantee future results")
print("  • Always consult financial advisors for investment decisions")

print("\n✅ End-to-End Stock Price Prediction Complete!")
print("=" * 50)

# 💾 OPTIONAL: Save the model
print("\n💾 Saving the trained model...")
model.save('adaniports_lstm_model.h5')
print("✅ Model saved as 'adaniports_lstm_model.h5'")

# Save the scaler for future use
import joblib
joblib.dump(scaler, 'adaniports_scaler.pkl')
print("✅ Scaler saved as 'adaniports_scaler.pkl'")
```
